```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

# Task
Data: [5.2, 5.4, 5.1, 5.6, 5.3, 5.8]<br>
Hypothesized mean: 5.5 <br>

Tasks:<br>
- Calculate the t-statistic and p-value.<br>
- Interpret and visualize the result at a 0.05 significance level.<br>


```python
observations = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])
expected_mean = 5.5
alpha = 0.05

test_stat, prob_value = stats.ttest_1samp(observations, expected_mean)

print(f"Test Statistic: {test_stat}")
print(f"P-value: {prob_value}")

if prob_value < alpha:
    print("Null hypothesis rejected. The average of the sample significantly differs from the expected mean.")
else:
    print("Null hypothesis not rejected. Insufficient evidence to state that the sample mean is different from the expected mean.")

plt.figure()
plt.bar(1, np.mean(observations), color="palegreen", label='Sample Average')
plt.errorbar(1, np.mean(observations), yerr=np.std(observations), fmt='k', linewidth=1.5)
plt.axhline(y=expected_mean, color='r', linestyle='--', linewidth=1.5, label='Expected Mean')
plt.title('Comparison of Sample Average and Expected Mean')
plt.xlim(0.5, 1.5)
plt.xlabel('Data Sample')
plt.ylabel('Mean Value')
plt.legend()
plt.grid(True)
plt.show()
```
