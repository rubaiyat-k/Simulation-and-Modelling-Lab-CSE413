### **Task-1:** Generate two vectors with 15 random floats each, plot them on the same graph, label the axes, and add a title.

```python
import numpy as np
import matplotlib.pyplot as plt

matrix1 = np.random.randint(4, 37, size=(4, 4))
matrix2 = np.random.randint(1, 29, size=(4, 4))

vector1 = matrix1.flatten()
vector2 = matrix2.flatten()

idx = np.arange(len(vector1))

plt.plot(idx, vector1, marker='D', linestyle='-', color='b', label="Set 1")
plt.plot(idx, vector2, marker='^', linestyle='--', color='r', label="Set 2")

plt.xlabel('Element Index')
plt.ylabel('Random Values')
plt.title('Randomized Data Sets')

plt.legend()
plt.show()
```

### **Task-2:** Create a 4x4 matrix with random values, visualize it as a heatmap, and label the rows and columns.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

grid = np.random.randint(4, 28, size=(4, 4))

rows = ['Top 1', 'Top 2', 'Top 3', 'Top 4']
cols = ['Left 1', 'Left 2', 'Left 3', 'Left 4']

sns.heatmap(grid, annot=True, cmap='coolwarm', linewidths=0.5)

plt.xticks(np.arange(len(cols)), cols)
plt.yticks(np.arange(len(rows)), rows)

plt.title('Heatmap Representation of Matrix')
plt.show()
```

### **Task-3:** Generate two 4x4 matrices, perform addition, subtraction, and multiplication, visualize the results using bar plots, and discuss the effects.

```python
import numpy as np
import matplotlib.pyplot as plt

M1 = np.random.randint(1, 10, (4, 4))
M2 = np.random.randint(1, 10, (4, 4))

sum_result = M1 + M2
diff_result = M1 - M2
product_result = np.dot(M1, M2)

print('Matrix Sum: \n', sum_result)
print('Matrix Difference: \n', diff_result)
print('Matrix Product: \n', product_result)

sum_values = sum_result.flatten()
diff_values = diff_result.flatten()
prod_values = product_result.flatten()

fig, bars = plt.subplots(1, 3, figsize=(20, 5))

bars[0].bar(range(len(sum_values)), sum_values, color='cyan')
bars[0].set_title("M1 + M2")
bars[0].set_ylim([0, 40])

bars[1].bar(range(len(diff_values)), diff_values, color='magenta')
bars[1].set_title("M1 - M2")
bars[1].set_ylim([-20, 20])

bars[2].bar(range(len(prod_values)), prod_values, color='navy')
bars[2].set_title("M1 * M2")
bars[2].set_ylim([0, 200])

plt.tight_layout()
plt.show()
```

#### **Discussion**
**Matrix Addition:** In the plot of the addition, we can see that each element is simply the sum of the corresponding elements of the two matrices. Since the numbers range from 1 to 9, the result will have values between 2 and 18.

**Matrix Subtraction:** The subtraction plot shows the difference between corresponding elements. The results can be both positive and negative depending on the relative values of A and B.

**Matrix Multiplication:** Matrix multiplication results in a more complex pattern compared to addition and subtraction. The values can vary more dramatically since each element is a sum of products of corresponding elements from rows of A and columns of B. This could lead to much larger values, especially if the original matrices contain large numbers.
